```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LOADING THE DATASET
data = pd.read_csv(R"C:\Users\Lenovo Laptop\Desktop\Unified-INT\FDI data (1).csv")
data
```

```python
# Display the first few rows to understand the structure
print("First few rows of the dataset:")
print(data.head())
```

```python
# Get basic statistics and information about the data
print("\nBasic Information:")
print(data.info())
print("\nBasic Statistics:")
print(data.describe())
```

```python
# Data Cleaning: Handle missing values
print("\nHandling missing values:")
data = data.fillna(0)  # Example: Filling missing values with 0
print(data.info())
```

```python
# Data Cleaning: Check for duplicate rows
print("\nChecking for duplicate rows:")
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Yearly growth rates calculation (example)
print("\nYearly growth rates:")
data_growth = data.set_index('Sector')
data_growth = data_growth.pct_change(axis='columns') * 100  # Percentage change year-over-year
print(data_growth.head())
```

```python
# Visualization: Yearly trend of FDI
sector_totals = data.set_index('Sector').sum(axis=1)
top_sectors = sector_totals.nlargest(5).index
data_top_sectors = data[data['Sector'].isin(top_sectors)]
plt.figure(figsize=(15, 7))
sns.lineplot(data=data_top_sectors.set_index('Sector').T)
plt.title('FDI Trends Over the Years (Top 5 Sectors)')
plt.xlabel('Year')
plt.ylabel('FDI Value')
plt.legend(title='Sector', bbox_to_anchor=(0.05, 1), loc='upper left')
plt.show()
```

```python
# Sector-wise analysis
print("\nSector-wise analysis:")
sector_totals = data.groupby('Sector').sum()
print(sector_totals)
```

```python
# Visualization: Sector-wise FDI distribution
plt.figure(figsize=(10, 8))
sector_totals.plot(kind='bar', stacked=True)
plt.title('Sector-wise FDI Distribution')
plt.xlabel('Sector')
plt.ylabel('Total FDI')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
